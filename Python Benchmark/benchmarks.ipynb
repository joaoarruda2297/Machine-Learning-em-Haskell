# Bibliotecas

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

# Carregando dataset e relizando divisões

```python
dataset = pd.read_csv('data/breast-cancer.csv')
dataset
```

```python
X = dataset[ [col for col in dataset.columns if (col != 'id' and col != 'diagnosis') ] ]
y = dataset[['diagnosis']]
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 42, shuffle= True)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.to_csv("outputs/X_train.csv", index = False)
X_test.to_csv("outputs/X_test.csv", index = False)
y_train.to_csv("outputs/y_train.csv", index = False)
y_test.to_csv("outputs/y_test.csv", index = False)
```

# Carregando modelos e realizando benchmarks

Convertendo para numpy array

```python
X_train = X_train.values
y_train = y_train['diagnosis'].values
X_test = X_test.values
y_test = y_test['diagnosis'].values
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model_knn = KNeighborsClassifier(n_neighbors= 5)
model_knn.fit(X_train, y_train)
predicted_knn = model_knn.predict(X_test)
```

```python
model_naive = GaussianNB()
model_naive.fit(X_train, y_train)
predicted_naive = model_naive.predict(X_test)
```

```python
model_tree = DecisionTreeClassifier(criterion= 'gini')
model_tree.fit(X_train, y_train)
predicted_tree = model_tree.predict(X_test)
```

# Avaliando predições

```python
acc_knn = accuracy_score(y_test, predicted_knn)
prec_knn = precision_score(y_test, predicted_knn, average="binary", pos_label="M")
recall_knn = recall_score(y_test, predicted_knn, average="binary", pos_label="M")
f1_knn = f1_score(y_test, predicted_knn, average="binary", pos_label="M")

acc_naive = accuracy_score(y_test, predicted_naive)
prec_naive = precision_score(y_test, predicted_naive, average="binary", pos_label="M")
recall_naive = recall_score(y_test, predicted_naive, average="binary", pos_label="M")
f1_naive = f1_score(y_test, predicted_naive, average="binary", pos_label="M")

acc_tree = accuracy_score(y_test, predicted_tree)
prec_tree = precision_score(y_test, predicted_tree, average="binary", pos_label="M")
recall_tree = recall_score(y_test, predicted_tree, average="binary", pos_label="M")
f1_tree = f1_score(y_test, predicted_tree, average="binary", pos_label="M")

print(f"KNN \nAcuracia:\t{acc_knn} \nPrecisao: \t{prec_knn} \nRecall: \t{recall_knn} \nF1 Score: \t{f1_knn}\n")
print(f"Naive Bayes \nAcuracia:\t{acc_naive} \nPrecisao: \t{prec_naive} \nRecall: \t{recall_naive} \nF1 Score: \t{f1_naive}\n")
print(f"Árvore Decisão \nAcuracia:\t{acc_tree} \nPrecisao: \t{prec_tree} \nRecall: \t{recall_tree} \nF1 Score: \t{f1_tree}")
```
